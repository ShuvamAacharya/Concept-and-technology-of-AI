```python
temperatures = [8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
                16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
                13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
                7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
                16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5]

cold = []
mild = []
comfort = []

for temp in temperatures:
    if temp < 10:
        cold.append(temp)
    elif 10 <= temp <= 15:
        mild.append(temp)
    elif 15 <= temp < 20:
        comfort.append(temp)

print("Cold:", cold)
print("Mild:", mild)
print("Comfort:", comfort)
```

task2

```python
cold_temperature_count =len(cold)
mild_temperature_count =len(mild)
comfort_temperature_count =len(comfort)

print(f"It was cold {cold_temperature_count} times.")
print(f"it was mild {mild_temperature_count} times.")
print(f"it was comfort {comfort_temperature_count} times.")
```

```python
temperatures_fahrenheit=[]
for celsius_temperature in temperatures:
    fahrenheit_temperature=(celsius_temperature *9/5)+32
    temperatures_fahrenheit.append(fahrenheit_temperature)
print("Temperature in fahrenheit:",temperatures_fahrenheit)
```
